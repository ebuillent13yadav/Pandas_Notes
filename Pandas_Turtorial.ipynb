# The isin() Function

Find the most destructive death over batsman in the history of IPL<br>
Strike rate = (Number of runs/Number of balls)*100<br>
Min batsman 200 balls in over 16-20

```python
new_deliveries = deliveries[deliveries['over'] > 15]
```

```python
# Fulfulling the minimum criteria
batsman = new_deliveries.groupby('batsman')['batsman_runs'].count()
indexes = batsman[batsman > 200].index.tolist()
```

```python
cri_deliveries = new_deliveries[new_deliveries['batsman'].isin(indexes)]
```

```python
runs = cri_deliveries.groupby('batsman')['batsman_runs'].sum()
balls = cri_deliveries.groupby('batsman')['batsman_runs'].count()
```

```python
((runs / balls) * 100).sort_values(ascending=False).head(1)
```

# The merge Function

## Find the batsman with most runs in a season

```python
new = deliveries.merge(data, left_on='match_id', right_on='id')
```

```python
batsman = new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
# reset_index changes a groupby object to dataframe
```

```python
batsman.drop_duplicates(subset='season',keep='first').sort_values('season')
```
